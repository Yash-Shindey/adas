```python
import pandas as pd

# Load the LIDAR data from the CSV file
lidar_data = pd.read_csv('../data/lidar_data.csv')

# Display the first few rows of the data
print(lidar_data.head())
```

```python
# Define a threshold for "nearby" obstacles
distance_threshold = 30  # in meters

# Filter the data to find obstacles within the threshold distance
nearby_obstacles = lidar_data[lidar_data['distance'] <= distance_threshold]

# Display the nearby obstacles
print("Nearby Obstacles:")
print(nearby_obstacles)
```

```python
import matplotlib.pyplot as plt

# Plot all data points
plt.scatter(lidar_data['x'], lidar_data['y'], c='blue', label='All data')

# Plot only the nearby obstacles
plt.scatter(nearby_obstacles['x'], nearby_obstacles['y'], c='red', label='Nearby Obstacles')

plt.xlabel('X Position (meters)')
plt.ylabel('Y Position (meters)')
plt.title('LIDAR Data: Obstacle Detection')
plt.legend()
plt.show()
```
